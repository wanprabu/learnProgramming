# Ordinary Least Squares (OLS)
Least squares is a standard tool in econometrics to assess the relationships between different variables. The OLS basically derives the basic regression model

Reference
https://www.r-econometrics.com/methods/ols/

## Arificial Example
The following model is used to create a sample 
$$y_i = 40 + 0.5x_i + e_i$$

where $e_i = N(0,4), x_i = U(1, 40)$

```r
# Reset random number generator for reproducibility
set.seed(1234)

# Total number of observations
N <- 50

# Simulate variable x from a uniform distribution
x <- runif(N, 1, 40)

# Simulate y
y <- 40 + .5 * x + rnorm(N, 0, 2)

# Note that the 'rnorm' function requires the standard deviation instead
# of the variance of the distribution. So, we have to enter 2 in order
# to draw from a normal distribution with variance 4.

# Store data in a data frame
ols_data <- data.frame(x, y)
```

```r
# Estimate the model and same the results in object "ols"
ols <- lm(y ~ x, data = ols_data) 

# Print the result in the console
summary(ols)
```

```r
library(ggplot2)

ggplot(ols_data, aes(x = x, y = y)) +
  geom_point() +
  geom_smooth(method = "lm", se = FALSE) +
  scale_x_continuous(limits = c(0, 45), expand = c(0, 0)) +
  theme_bw()
```

## Weight curve for salmon
The model for weight follows the biological growth as follow:
$$w_t = 5.72t^2 - 2.08t^3$$,
where t = 32.81 - 12.52

```r
# Reset random number generator for reproducibility
set.seed(1234)

# Total number of observations
N <- 100

# Simulate variable x from a uniform distribution
t1 <- runif(N, 1, 32.81)
t2 <- -12.52
t <- seq(1, 2*12)

# Simulate y
w <- 5.72*t**2 - 2.08 * t**3

# Note that the 'rnorm' function requires the standard deviation instead
# of the variance of the distribution. So, we have to enter 2 in order
# to draw from a normal distribution with variance 4.

# Store data in a data frame
weight_data <- data.frame(t, w)
weight_data
```

```r
# Estimate the model and same the results in object "ols"
weight <- lm(w ~ t, data = weight_data) 

# Print the result in the console
summary(weight)
```

```r
library(ggplot2)

ggplot(weight_data, aes(x = t, y = w)) +
  geom_point() +
  geom_smooth(method = "lm", se = FALSE) +
  scale_x_continuous(limits = c(0, 45), expand = c(0, 0)) +
  theme_bw()
```
